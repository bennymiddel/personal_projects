```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
p = np.linspace(0,1, 100)

A_pay = 30
A_loss = 20
B_pay = 100
B_loss = 40
```

```python
A = (p*A_pay) - ((1-p)*A_loss)
B = (p*B_pay) - ((1-p)*B_loss)
```

```python
plt.figure(dpi = 300, figsize = (5,5))
plt.plot(A, color = 'red', label = 'A')
plt.plot(B, color = 'blue', label = 'B')
plt.vlines(x=50, ymin =-40, ymax =100, linestyles= '--')
plt.legend()
plt.xlabel('Chance of Pay')
plt.ylabel('expected value')
plt.show()
```
