```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
p = np.linspace(0,1, 100)

A_pay = 10
A_loss = 20
B_pay = 60
B_loss = 40
```

```python
A = (p*A_pay) - ((1-p)*A_loss)
B = (p*B_pay) - ((1-p)*B_loss)
```

```python
threshold = (A_loss-B_loss)/(A_pay+A_loss-B_pay-B_loss)
```

```python
plt.figure(dpi = 100, figsize = (5,5))
plt.plot(p, A, color = 'red', label = 'Scenario 1')
plt.plot(p, B, color = 'blue', label = 'Scenario 2')
plt.axvline(x=0.5, linestyle= '--', color ='gray', linewidth = 0.6)
plt.axvspan(xmin=0.0, xmax = threshold, label = 'Ayo and Benedikt', alpha = 0.3)
plt.axvspan(xmin=threshold, xmax = 1, label = 'Cybersecurity class without Ayo and Benedikt' , alpha = 0.3, color = 'black')
plt.axvline(x=threshold, linestyle = '--', color = 'darkred', label = f'Threshold: {round(threshold, 2)}', linewidth = 0.6)
plt.legend()
plt.xlabel('Chance of positive pay')
plt.ylabel('expected value in M$')
plt.show()
```

```python
def is_leap(year):
    leap = False
    
    if year%4==0:
        if year%100==0:
            if year%400==0:
                leap = True
            else:
                leap = False
        else:
            leap = True
    else:
        leap = False

       
    
    return leap

```

```python
def is_leap_2(year):
    return (year % 4 == 0) and (year % 100 != 0 or year % 400 == 0)
```

```python
input = 2096
print(is_leap(input))
print(is_leap_2(input))
```
